```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Proga\AML\PPG_dataset\dataset.csv")

# Подсчет количества встречаемости каждого age_group
age_group_counts = df['age_group'].value_counts().sort_index()

# Вывод результатов
print(age_group_counts)
```

```python
plt.figure(figsize=(10, 6))
age_group_counts.plot(kind='bar', color='skyblue')
plt.title('Количество встречаемости age_group')
plt.xlabel('Age Group')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Отображение графика
plt.show()
```

```python
df_limited = df.groupby('age_group').head(50)

# Подсчет количества встречаемости каждого age_group
age_group_counts = df_limited['age_group'].value_counts().sort_index()

# Построение графика
plt.figure(figsize=(10, 6))
age_group_counts.plot(kind='bar', color='skyblue')
plt.title('Количество встречаемости age_group (ограничено до 50 записей)')
plt.xlabel('Age Group')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Отображение графика
plt.show()
```
